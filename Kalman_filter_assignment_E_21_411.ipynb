# Q. Analytical Derivation

Consider the filter model
\begin{align*}
x^{-}_k &= A_{k-1}\,x^{+}_{k-1} + G_{k-1}\,w_{k-1}, \\
y^{-}_k &= H_k\,x^{-}_k + z_k,
\end{align*}
where
$$x^{+}_{k-1} \sim \mathscr{N}(m_{k-1}, P_{k-1}),$$

Answer the followings:

1. Show that
$$x_k^- \sim \mathscr{N}(m_k^-, P_k^-),$$
where
$$m_k^- = A_{k-1}m_{k-1},$$
and
$$
P_k^- = A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T.
$$

2. Show that
$$y_k^- \sim \mathscr{N}(H_k m_k^-, H_k P_k^- H_k^T + \Sigma_m).$$

3. Show that
$$\begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix} \sim \mathscr{N}\left(
\begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},
\begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}
\right).$$

4. Show that
$$x^{+}_k\triangleq (x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}) \sim \mathscr{N}(m_k, P_k),$$
where the updated mean $m_k$ and updated covariance $P_k$ are given by:
\begin{align*}
K_k &\triangleq P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}, \\
m_k &= m_k^- + K_k (y^{\mathrm{obs}}_{k} - H_k m_k^-), \\
P_k &= (I - K_k H_k) P_k^-.
\end{align*}

5. Find
$\mathbb{E}[x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}]$  and
$\text{Var}(x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k})$.

### Solution: Analytical Derivation

**1. Prediction Step**
Taking the expectation of the state equation yields the predicted mean:
$$m_k^- \triangleq \mathbb{E}[x^{-}_k] = A_{k-1}\mathbb{E}[x^{+}_{k-1}] + G_{k-1}\mathbb{E}[w_{k-1}] = A_{k-1}m_{k-1},$$
since $\mathbb{E}[w_{k-1}] = 0$.

The predicted covariance $P_k^-$ is computed by applying the variance operator:
\begin{align*}
P_k^- &\triangleq \text{Var}(x^{-}_k) \\
&= A_{k-1}\text{Var}(x^{+}_{k-1})A_{k-1}^T + G_{k-1}\text{Var}(w_{k-1})G_{k-1}^T \\
&= A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T.
\end{align*}
Thus, $x_k^- \sim \mathscr{N}(m_k^-, P_k^-)$.

**2. Measurement Prediction**
Based on our prior state belief, the expected value of the measurement is:
$$\mathbb{E}[y^{-}_k] = H_k \mathbb{E}[x_k^-] + \mathbb{E}[z_k] = H_k m_k^-$$
and its variance is:
$$\text{Var}(y^{-}_k) = H_k \text{Var}(x_k^-) H_k^T + \text{Var}(z_k) = H_k P_k^- H_k^T + \Sigma_m$$
Thus, $y_k^- \sim \mathscr{N}(H_k m_k^-, H_k P_k^- H_k^T + \Sigma_m)$.

**3. Joint Prior Distribution**
The cross-covariance is:
$$\text{Cov}(x_k^-, y^{-}_k) = \text{Cov}(x_k^-, H_k x_k^- + z_k) = P_k^- H_k^T.$$
Therefore, the joint distribution is:
$$\begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix} \sim \mathscr{N}\left(
\begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},
\begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}
\right).$$

**4. & 5. Measurement Update (Correction Step) & Expectations**
By applying standard conditioning properties of multivariate normal distributions on the joint distribution at $y_k = y^{\mathrm{obs}}_{k}$, we extract the conditional mean and covariance:
$$\mathbb{E}[x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}] = m_k^- + P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1} (y^{\mathrm{obs}}_{k} - H_k m_k^-)$$
$$\text{Var}(x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}) = P_k^- - P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1} H_k P_k^-$$
Defining the Kalman Gain $K_k \triangleq P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}$, we get the updated mean $m_k = m_k^- + K_k (y^{\mathrm{obs}}_{k} - H_k m_k^-)$ and covariance $P_k = (I - K_k H_k) P_k^-$. These exactly correspond to the required expressions for the updated mean and covariance.


# Q. 1-D Example

Consider the scalar linear-Gaussian filter model
\begin{aligned}
x^-_k &= a\,x^+_{k-1} + w_{k-1},\qquad w_{k-1}\sim\mathscr N(0,q),\\
y^-_k &= h\,x^-_k + z_k,\qquad\;\;\;\; z_k\sim\mathscr N(0,r),
\end{aligned}
with prior $x_0\sim \mathscr{N}(m_0,P_0)$. Define $Y_{k}=\{y^{\mathrm{obs}}_1,\dots,y^{\mathrm{obs}}_{k}\}$.

Answer the following:

1. Show that
\begin{aligned}
m_k^- &= a\,m_{k-1},\\
P_k^- &= a^2 P_{k-1} + q.
\end{aligned}

2. Show that
\begin{aligned}
m_k &= m_k^- + K_k\,v_k
= m_k^- + \frac{P_k^- h}{S_k}\bigl(y^{\mathrm{obs}}_k - h\,m_k^-\bigr),\\
P_k &= (1 - K_k h)\,P_k^-
= \Bigl(1 - \frac{P_k^- h^2}{S_k}\Bigr) P_k^- .
\end{aligned}

3. Show that the predictive measurement distribution (before seeing $y_k$):
\begin{align*}
p(y^-_k | Y_{k-1})=\mathscr N\bigl(h\,m_k^-,\; h^2 P_k^- + r\bigr).
\end{align*}

4. Show that the posterior-predictive measurement distribution (after filtering on $y_k$):
\begin{align*}
p(y^-_k\mid Y_k)=\mathscr N\bigl(h\,m_k,\; h^2 P_k + r\bigr).
\end{align*}

5. For suitable numerical values for $a,q,h,r$ animate the above to prior distribution and the posterior distribution.

### Solution: 1-D Example

**1. Prediction**
\begin{aligned}
m_k^- &= \mathbb{E}[a\,x^+_{k-1} + w_{k-1}] = a\,m_{k-1},\\
P_k^- &= \text{Var}(a\,x^+_{k-1} + w_{k-1}) = a^2 P_{k-1} + q.
\end{aligned}

**2. Update**
The innovation variance is $S_k = h^2 P_k^- + r$. The Kalman Gain is $K_k = \frac{P_k^-\,h}{S_k}$.
\begin{aligned}
m_k &= m_k^- + K_k\,v_k = m_k^- + \frac{P_k^- h}{S_k}\bigl(y^{\mathrm{obs}}_k - h\,m_k^-\bigr),\\
P_k &= (1 - K_k h)\,P_k^- = \Bigl(1 - \frac{P_k^- h^2}{S_k}\Bigr) P_k^- .
\end{aligned}

**3. & 4. Predictive Distributions**
Prior to observing $y_k$:
\begin{align*}
p(y^-_k \mid Y_{k-1}) &= \mathscr N\bigl(h\,m_k^-,\; h^2 P_k^- + r\bigr).
\end{align*}
After observing $y_k$ (posterior-predictive):
\begin{align*}
p(y^-_k \mid Y_k) &= \mathscr N\bigl(h\,m_k,\; h^2 P_k + r\bigr).
\end{align*}

# Q. 2D-Position Estimation

## Part A
Consider a two-dimensional constant-velocity dynamical system. The hidden state at time step $k$ is

$$
x_k =
\begin{bmatrix}
p_x(k)\\
p_y(k)\\
v_x(k)\\
v_y(k)
\end{bmatrix},
$$

where $(p_x(k),p_y(k))$ denote the position components and $(v_x(k),v_y(k))$ denote the velocity components.

The measurement consists only of the two position components:

$$
y_k =
\begin{bmatrix}
p_x^{\mathrm{meas}}(k)\\
p_y^{\mathrm{meas}}(k)
\end{bmatrix}.
$$

The linear Gaussian state-space filter model is

$$
x^-_k = A x^+_{k-1} + G w_{k-1},
$$

$$
y^-_k = Hx^+_k + z_k,
$$

where

$$
w_{k-1} \sim \mathscr{N}(0,\Sigma_p),
\qquad
z_k \sim \mathscr{N}(0,\Sigma_m).
$$

The process noise sequence $w_k$, measurement noise sequence $z_k$, and the initial state are assumed mutually independent.

Show that:

$$
A =
\begin{bmatrix}
1 & 0 & \Delta t & 0\\
0 & 1 & 0 & \Delta t\\
0 & 0 & 1 & 0\\
0 & 0 & 0 & 1
\end{bmatrix},
$$

$$
H =
\begin{bmatrix}
1 & 0 & 0 & 0\\
0 & 1 & 0 & 0
\end{bmatrix},
$$

and

$$
G =
\begin{bmatrix}
\frac{1}{2}\Delta t^2 & 0\\
0 & \frac{1}{2}\Delta t^2\\
\Delta t & 0\\
0 & \Delta t
\end{bmatrix}.
$$

### Solution: 2D-Position Estimation - Part A

Assuming a sampling interval $\Delta t$, the constant-velocity kinematic equations are:
$$
p_x(k) = p_x(k-1) + \Delta t\,v_x(k-1),
$$
$$
p_y(k) = p_y(k-1) + \Delta t\,v_y(k-1),
$$
$$
v_x(k) = v_x(k-1),
$$
$$
v_y(k) = v_y(k-1).
$$
In matrix form, mapping $x_{k-1}$ to $x_k$ gives the transition matrix $A$:
$$
A =
\begin{bmatrix}
1 & 0 & \Delta t & 0\\
0 & 1 & 0 & \Delta t\\
0 & 0 & 1 & 0\\
0 & 0 & 0 & 1
\end{bmatrix}.
$$

Since the measurement isolates the first two position components, the observation matrix $H$ is:
$$
H =
\begin{bmatrix}
1 & 0 & 0 & 0\\
0 & 1 & 0 & 0
\end{bmatrix}.
$$

A random unmodeled acceleration $a(k-1) = [a_x(k-1), a_y(k-1)]^T$ operates over the interval $\Delta t$. Its effect on position is $\frac{1}{2}\Delta t^2 a(k-1)$ and on velocity is $\Delta t\,a(k-1)$. Consequently, the matrix $G$ mapping these into the state space is:
$$
G =
\begin{bmatrix}
\frac{1}{2}\Delta t^2 & 0\\
0 & \frac{1}{2}\Delta t^2\\
\Delta t & 0\\
0 & \Delta t
\end{bmatrix}.
$$

## Part B
Develop a python based method to filter a sequence of noisy GPS obtained position meaurements.

```python
import numpy as np
import sims

def make_cv2d(dt=0.1, q=1e-2, r=0.5, x0=(0, 0, 1, 0.5), seed=0):
    A = np.array([
        [1, 0, dt,  0],
        [0, 1,  0, dt],
        [0, 0,  1,  0],
        [0, 0,  0,  1],
    ])
    H = np.array([
        [1, 0, 0, 0],
        [0, 1, 0, 0],
    ])
    R = np.eye(2) * (r**2)
    G = np.array([
        [0.5*dt*dt, 0.0      ],
        [0.0      , 0.5*dt*dt],
        [dt       , 0.0      ],
        [0.0      , dt       ],
    ])
    Sigma_p = np.eye(2) * (q**2)

    return sims.LinearGaussianSystemSyms(
        A=A, H=H, Sigma_p=Sigma_p, Sigma_m=R,
        x0=np.asarray(x0, float),
        rng=np.random.default_rng(seed),
        G=G
    )

# Initialize system
sys_2d = make_cv2d(dt=0.1, q=1e-2, r=0.5, x0=(0, 0, 0.5, -0.2), seed=7)

# Simulate noisy GPS observations (Y_gps)
X_true, Y_gps = sys_2d.simulate(T=300)

# Filter the sequence of noisy GPS measurements and plot the result
M, Yhat = sys_2d.filter_with_kf_and_plot(
    T=300,
    Y=Y_gps,
    m0=np.array([0, 0, 0, 0]),
    P0=np.eye(4)*100,
    component_labels=["pos_x","pos_y"],
    show=True
)
```
