#Q3) Transformer Failiures

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$
P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.
$$

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.
2. Define $X$ as a random variable on this space.
3. Determine the probability mass function of $X$.
4. Compute the expectation $E[X]$.
5. Compute the variance $\mathrm{Var}(X)$.
6. Find $P(X\ge 1)$ and $P(X\ge 2)$.
7. Briefly explain the practical meaning of the expectation and variance in this setting.



```python
import numpy as np
from IPython.display import display, Math

# -----------------------------
# Probability Distribution
# -----------------------------

X = np.array([0,1,2,3])

P = np.array([0.50,0.30,0.15,0.05])

# -----------------------------
# Calculations
# -----------------------------

EX = np.sum(X*P)

EX2 = np.sum((X**2)*P)

VarX = EX2 - EX**2

P_ge_1 = np.sum(P[X>=1])

P_ge_2 = np.sum(P[X>=2])

# -----------------------------
# Answers
# -----------------------------
print('Q1')
# Q1
display(Math(r"\Omega=\{0,1,2,3\}"))

display(Math(r"F=2^\Omega"))
print('Q2')
# Q2
display(Math(r"X:\Omega\to\mathbb{R}"))

display(Math(r"X(\omega)=\omega"))
print('Q3')
# Q3
display(Math(r"""
p_X(x)=
\begin{cases}
0.50,& x=0\\
0.30,& x=1\\
0.15,& x=2\\
0.05,& x=3\\
0,& \text{otherwise}
\end{cases}
"""))
print('Q4')
# Q4
display(Math(r"""
E[X]=\sum xP(X=x)
"""))

display(Math(r"""
E[X]=0(0.50)+1(0.30)+2(0.15)+3(0.05)
"""))

display(Math(rf"""
E[X]={EX}
"""))
print('Q5')
# Q5
display(Math(r"""
E[X^2]=0^2(0.50)+1^2(0.30)+2^2(0.15)+3^2(0.05)
"""))

display(Math(rf"""
E[X^2]={EX2}
"""))

display(Math(r"""
Var(X)=E[X^2]-(E[X])^2
"""))

display(Math(rf"""
Var(X)={VarX}
"""))
print('Q6')
# Q6
display(Math(r"""
P(X\geq1)=P(X=1)+P(X=2)+P(X=3)
"""))

display(Math(rf"""
P(X\geq1)={P_ge_1}
"""))

display(Math(r"""
P(X\geq2)=P(X=2)+P(X=3)
"""))

display(Math(rf"""
P(X\geq2)={P_ge_2}
"""))
print('Q7')
# Q7
display(Math(r"""
\text{Expectation represents the average number of transformer failures per day.}
"""))

display(Math(r"""
\text{Variance measures how much the failures fluctuate from day to day.}
"""))
```

# Q4) Probability of Defective Items

In a manufacturing process, the proportion of defective items in a large batch produced on a given day is modeled by a continuous random variable $X$. Since $X$ represents a proportion, it takes values in the interval $[0,1]$. Suppose that $X$ has a Beta distribution with parameters $\alpha=2$ and $\beta=18$, so that its density is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x^{1}(1-x)^{17}, & 0<x<1,\\\
0, & \text{otherwise},
\end{cases}
$$

where $B(2,18)$ is the Beta function.

Answer the following:

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.

2. Define $X$ as a continuous random variable on this space.

3. Write down the probability density function of $X$.

4. Verify that $f$ is a valid probability density function.

5. Compute the expectation $E[X]$.

6. Compute the variance $\mathrm{Var}(X)$.

7. Find the probability that the defective proportion is at most $0.10$, that is, compute

$$
P(X\le 0.10).
$$

You may leave your answer in Beta integral form.

8. Find the probability that the defective proportion lies between $0.05$ and $0.15$, that is, compute

$$
P(0.05\le X\le 0.15).
$$

You may leave your answer in integral form.

9. Briefly explain why a Beta distribution is a reasonable model for this situation.

10. Interpret the expectation and variance in the context of production quality.


```python
import numpy as np
from IPython.display import display, Math

# Parameters of Beta distribution
alpha = 2
beta = 18

# -----------------------------
# Expectation and Variance
# -----------------------------

EX = alpha/(alpha+beta)

VarX = (alpha*beta)/(((alpha+beta)**2)*(alpha+beta+1))

# -----------------------------
# Answers
# -----------------------------

print('Q1')
display(Math(r"\Omega=[0,1]"))

display(Math(r"F=\mathcal{B}([0,1])"))

print('Q2')
display(Math(r"X:\Omega\to\mathbb{R}"))

display(Math(r"X(\omega)=\omega"))

print('Q3')
display(Math(r"""
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x(1-x)^{17}, & 0<x<1\\
0, & \text{otherwise}
\end{cases}
"""))

print('Q4')
display(Math(r"""
\int_0^1 f(x)\,dx = 1
"""))

display(Math(r"""
\therefore f(x)\text{ is a valid probability density function.}
"""))

print('Q5')
display(Math(r"""
E[X]=\frac{\alpha}{\alpha+\beta}
"""))

display(Math(rf"""
E[X]=\frac{{2}}{{20}}={EX}
"""))

print('Q6')
display(Math(r"""
Var(X)=
\frac{\alpha\beta}
{(\alpha+\beta)^2(\alpha+\beta+1)}
"""))

display(Math(rf"""
Var(X)={VarX}
"""))

print('Q7')
display(Math(r"""
P(X\leq0.10)=
\int_0^{0.10}
\frac{1}{B(2,18)}x(1-x)^{17}\,dx
"""))

print('Q8')
display(Math(r"""
P(0.05\leq X\leq0.15)=
\int_{0.05}^{0.15}
\frac{1}{B(2,18)}x(1-x)^{17}\,dx
"""))

print('Q9')
display(Math(r"""
\text{A Beta distribution is suitable because proportions are bounded between }0\text{ and }1.
"""))

print("Q10")
display(Math(r"""
\text{Expectation represents the average defective proportion in production.}
"""))

display(Math(r"""
\text{Variance measures the variability of production quality between days.}
"""))

# -----------------------------
# Print Numerical Results
# -----------------------------

print("E[X] =", EX)

print("Var(X) =", VarX)
```

# Q5) Expected Project Completion Time

A construction company is estimating the time required to install a backup generator at a hospital. Let $X$ denote the number of hours needed to complete the installation. Based on expert judgment, the shortest possible time is $4$ hours, the longest possible time is $10$ hours, and the most likely completion time is $6$ hours. Assume that $X$ follows a Triangular distribution with parameters $a=4$, $b=10$, and $c=6$.

Answer the following:

1. Specify the support of $X$.

2. Write down the probability density function $f_X(x)$.

3. Verify that $f_X$ integrates to $1$ over its support.

4. Compute the expectation $E[X]$.

5. Compute the variance $\mathrm{Var}(X)$.

6. Interpret the values of $a$, $b$, and $c$ in the context of this problem.

7. Explain briefly why a Triangular distribution is a reasonable model in this setting.

8. Compute the probability that the installation takes at most $5$ hours, that is, compute

$$
P(X \leq 5).
$$

9. Compute the probability that the installation takes between $5$ and $8$ hours, that is, compute

$$
P(5 \leq X \leq 8).
$$

10. Briefly explain how this model could be used in project planning or sensitivity analysis.


```python
import numpy as np
from IPython.display import display, Math

# Parameters
a = 4
b = 10
c = 6

# -----------------------------
# Expectation and Variance
# -----------------------------

EX = (a+b+c)/3

VarX = (a**2 + b**2 + c**2 - a*b - a*c - b*c)/18

# -----------------------------
# Probabilities
# -----------------------------

# P(X<=5)
P1 = ((5-a)**2)/((b-a)*(c-a))

# CDF at x=8
F8 = 1 - ((b-8)**2)/((b-a)*(b-c))

# CDF at x=5
F5 = ((5-a)**2)/((b-a)*(c-a))

P2 = F8 - F5

# -----------------------------
# Answers
# -----------------------------
print('Q1')
# Q1
display(Math(r"4\leq X\leq10"))
print('Q2')
# Q2
display(Math(r"""
f_X(x)=
\begin{cases}
\dfrac{2(x-4)}{(10-4)(6-4)}, & 4\leq x\leq6\\
\dfrac{2(10-x)}{(10-4)(10-6)}, & 6<x\leq10\\
0, & \text{otherwise}
\end{cases}
"""))
print('Q3')
# Q3
display(Math(r"""
\int_4^{10} f_X(x)\,dx = 1
"""))

display(Math(r"""
\therefore f_X(x)\text{ is a valid probability density function.}
"""))
print('Q4')
# Q4
display(Math(r"""
E[X]=\frac{a+b+c}{3}
"""))

display(Math(rf"""
E[X]={EX}
"""))
print('Q5')
# Q5
display(Math(r"""
Var(X)=
\frac{a^2+b^2+c^2-ab-ac-bc}{18}
"""))

display(Math(rf"""
Var(X)={VarX}
"""))
print('Q6')
# Q6
display(Math(r"""
a=4 \text{ : minimum completion time}
"""))

display(Math(r"""
b=10 \text{ : maximum completion time}
"""))

display(Math(r"""
c=6 \text{ : most likely completion time}
"""))
print('Q7')
# Q7
display(Math(r"""
\text{Triangular distributions are useful when only minimum, maximum,}
"""))

display(Math(r"""
\text{and most likely values are known from expert judgment.}
"""))
print('Q8')
# Q8
display(Math(r"""
P(X\leq5)=
\frac{(5-4)^2}{(10-4)(6-4)}
"""))

display(Math(rf"""
P(X\leq5)={P1}
"""))
print('Q8')
# Q9
display(Math(r"""
P(5\leq X\leq8)=F(8)-F(5)
"""))

display(Math(rf"""
P(5\leq X\leq8)={P2}
"""))
print('Q10')
# Q10
display(Math(r"""
\text{This model helps estimate project duration uncertainty and scheduling risk.}
"""))

display(Math(r"""
\text{It can also support sensitivity analysis and resource planning.}
"""))

# -----------------------------
# Print Results
# -----------------------------

print("E[X] =", EX)

print("Var(X) =", VarX)

print("P(X<=5) =", P1)

print("P(5<=X<=8) =", P2)
```

#Q6) Expected Project Completion Time Using PERT

A project manager is estimating the time required to complete a critical software integration task. Let $X$ denote the number of days required to complete the task. Based on expert judgment, the shortest possible time is $a=8$ days, the longest possible time is $b=20$ days, and the most likely completion time is $c=11$ days. Assume that $X$ follows a PERT distribution with parameters $(a,b,c)$.

The PERT distribution is defined as a Beta distribution scaled to the interval $[a,b]$, with internal shape parameters

$$
\alpha = 1 + 4\left( \frac{c-a}{b-a} \right)
$$

and

$$
\beta = 1 + 4\left( \frac{b-c}{b-a} \right).
$$

Its probability density function is

$$
f(x)=\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}, \qquad a \leq x \leq b,
$$

where

$$
B(\alpha,\beta)=\int_0^1 t^{\alpha-1}(1-t)^{\beta-1},dt.
$$

Answer the following:

1. Compute the values of the shape parameters $\alpha$ and $\beta$.

2. Specify the support of $X$.

3. Write down the probability density function $f_X(x)$ for this model.

4. State the probability space $(\Omega,\mathfrak F,P)$ suitable for this continuous model.

5. Define $X$ as a random variable on this space.

6. Show that the expectation $E[X]$ is given by

$$
E[X]=\frac{a+4c+b}{6}.
$$

7. Compute the variance $\mathrm{Var}(X)$ using the approximation

$$
\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}.
$$

8. Compute the approximate standard deviation $\sigma$.

9. Briefly explain why a PERT distribution may be more appropriate than a Triangular distribution in this setting byt comparing the two results.

10. Interpret the practical meaning of the values $a$, $b$, $c$, $E[X]$, and $\sigma$ in the context of project planning.


```python
import numpy as np
from IPython.display import display, Math

# Parameters
a = 8
b = 20
c = 11



alpha = 1 + 4*((c-a)/(b-a))

beta = 1 + 4*((b-c)/(b-a))



EX = (a + 4*c + b)/6



VarX = ((EX-a)*(b-EX))/7

sigma = np.sqrt(VarX)

print('Q1')
# Q1
display(Math(r"""
\alpha=1+4\left(\frac{c-a}{b-a}\right)
"""))

display(Math(rf"""
\alpha={alpha}
"""))

display(Math(r"""
\beta=1+4\left(\frac{b-c}{b-a}\right)
"""))

display(Math(rf"""
\beta={beta}
"""))
print('Q2')
# Q2
display(Math(r"8\leq X\leq20"))
print('Q3')
# Q3
display(Math(r"""
f_X(x)=
\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}
{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}},
\quad a\leq x\leq b
"""))
print('Q4')
# Q4
display(Math(r"\Omega=[8,20]"))

display(Math(r"F=\mathcal{B}([8,20])"))
print('Q5')
# Q5
display(Math(r"X:\Omega\to\mathbb{R}"))

display(Math(r"X(\omega)=\omega"))
print('Q6')
# Q6
display(Math(r"""
E[X]=\frac{a+4c+b}{6}
"""))

display(Math(rf"""
E[X]={EX}
"""))
print('Q7')
# Q7
display(Math(r"""
Var(X)=
\frac{(E[X]-a)(b-E[X])}{7}
"""))

display(Math(rf"""
Var(X)={VarX}
"""))
print('Q8')
# Q8
display(Math(r"""
\sigma=\sqrt{Var(X)}
"""))

display(Math(rf"""
\sigma={sigma}
"""))
print('Q9')
# Q9
display(Math(r"""
\text{PERT distributions are smoother and place more weight near the most likely value.}
"""))

display(Math(r"""
\text{Triangular distributions are simpler but less realistic for uncertainty modeling.}
"""))
print('Q10')
# Q10
display(Math(r"""
a=8 \text{ : shortest possible completion time}
"""))

display(Math(r"""
b=20 \text{ : longest possible completion time}
"""))

display(Math(r"""
c=11 \text{ : most likely completion time}
"""))

display(Math(r"""
E[X] \text{ represents the expected project completion time.}
"""))

display(Math(r"""
\sigma \text{ measures uncertainty and variability in project duration.}
"""))


# Print Results


print("alpha =", alpha)

print("beta =", beta)

print("E[X] =", EX)

print("Var(X) =", VarX)

print("sigma =", sigma)
```

#Q7) If the PERT Expectation is to Match Our Intuition

A project manager models the duration $X$ of a task by a PERT distribution with parameters $a$, $b$, and $c$, where $a$ is the minimum time, $b$ is the maximum time, and $c$ is the most likely time.

Suppose the minimum time is $a=5$ days and the most likely time is $c=11$ days. If the expected time computed from the PERT model is also to be the same as the most likely time:

1. Determine the value of the maximum time $b$.

2. What happens to the variance?

```python
import numpy as np
from IPython.display import display, Math

# Given values
a = 5
c = 11

# --------------------------------
# Solve for b using
# E[X] = (a + 4c + b)/6
# and E[X] = c
# --------------------------------

b = 6*c - a - 4*c

# --------------------------------
# Expectation
# --------------------------------

EX = (a + 4*c + b)/6

# --------------------------------
# Variance
# --------------------------------

VarX = ((EX-a)*(b-EX))/7

# --------------------------------
# Answers
# --------------------------------

display(Math(r"""
E[X]=\frac{a+4c+b}{6}
"""))

display(Math(r"""
\text{Since } E[X]=c,
"""))

display(Math(r"""
c=\frac{a+4c+b}{6}
"""))

display(Math(r"""
6c=a+4c+b
"""))

display(Math(r"""
b=2c-a
"""))

display(Math(rf"""
b=2({c})-({a})
"""))

display(Math(rf"""
b={b}
"""))

# Variance
display(Math(r"""
Var(X)=
\frac{(E[X]-a)(b-E[X])}{7}
"""))

display(Math(rf"""
Var(X)=\frac{{({EX}-{a})({b}-{EX})}}{{7}}
"""))

display(Math(rf"""
Var(X)={VarX}
"""))

display(Math(r"""
\text{The variance remains positive, meaning uncertainty still exists in task duration.}
"""))

# --------------------------------
# Print Results
# --------------------------------

print("Maximum time b =", b)

print("Variance =", VarX)
```
