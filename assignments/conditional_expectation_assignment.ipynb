#Q1) Prediction of fraudulent transactions

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer's risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

3. Explain why $E[X\mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets $\{ \omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$.

4. Compute $E[X\mid Y]$ explicitly and show that

$$
E[X\mid Y]= \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}
$$

5. Verify the defining property of conditional expectation by checking that for every set $A\in\sigma(Y)$,

$$
\int_A E[X\mid Y]\,dP=\int_A X\,dP.
$$

6. Find the Conditional PMF $p_{X|Y}(x \mid y)$.

7. Verify the identity:
$$\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$$.

8. Calculate the **Mean Square Error (MSE)**

9. How would the **Mean Square Error (MSE)** change if the bank decided to ignore the customer risk class $Y$ and instead used the global average fraud rate ($E[X] = 0.30$) as a constant estimate for every transaction? Calculate this error and explain its relationship to the **Law of Total Variance**.

10. Using the joint distribution of $(X, Y)$ derived from the sample space:

    a) Compute the entropy $H(X)$, representing the bank's initial uncertainty regarding fraud.

    b) Compute the conditional entropy $H(X \mid Y)$, representing the uncertainty remaining after the customer's risk class is revealed.

    c) Calculate the Mutual Information $I(X; Y)$.

    d) Based on these values, what percentage of the total uncertainty about fraud is removed by knowing the customer's risk class?

```python


import math



Omega = ['w1', 'w2', 'w3', 'w4']

P = {
    'w1': 0.50,
    'w2': 0.10,
    'w3': 0.20,
    'w4': 0.20
}

# Fraud Indicator X
X = {
    'w1': 0,
    'w2': 1,
    'w3': 0,
    'w4': 1
}

# Risk Class Y
Y = {
    'w1': 0,
    'w2': 0,
    'w3': 1,
    'w4': 1
}

# ------------------------------------------------------------
# 2. Sigma Algebra Generated by Y
# ------------------------------------------------------------

sigma_Y = [
    set(),
    {'w1', 'w2'},
    {'w3', 'w4'},
    set(Omega)
]

print("Sigma Algebra Generated by Y:")
for s in sigma_Y:
    print(s)

# ------------------------------------------------------------
# 3. Compute Conditional Expectation E[X|Y]
# ------------------------------------------------------------

# P(Y=0)
P_Y0 = P['w1'] + P['w2']

# P(Y=1)
P_Y1 = P['w3'] + P['w4']

# E[X|Y=0]
EX_Y0 = P['w2'] / P_Y0

# E[X|Y=1]
EX_Y1 = P['w4'] / P_Y1

print("\nConditional Expectations:")
print("E[X|Y=0] =", EX_Y0)
print("E[X|Y=1] =", EX_Y1)

# Define E[X|Y](w)
EX_given_Y = {}

for w in Omega:
    if Y[w] == 0:
        EX_given_Y[w] = EX_Y0
    else:
        EX_given_Y[w] = EX_Y1

print("\nE[X|Y] on Omega:")
for w in Omega:
    print(f"{w}: {EX_given_Y[w]}")

# ------------------------------------------------------------
# 4. Verify Defining Property
# ------------------------------------------------------------

print("\nVerification of Conditional Expectation Property:")

for A in sigma_Y[1:3]:

    lhs = sum(EX_given_Y[w] * P[w] for w in A)
    rhs = sum(X[w] * P[w] for w in A)

    print(f"\nSet A = {A}")
    print("Integral of E[X|Y] over A =", lhs)
    print("Integral of X over A      =", rhs)

# ------------------------------------------------------------
# 5. Conditional PMF p(X|Y)
# ------------------------------------------------------------

print("\nConditional PMF:")

print("\nFor Y = 0")
print("P(X=0|Y=0) =", 1 - EX_Y0)
print("P(X=1|Y=0) =", EX_Y0)

print("\nFor Y = 1")
print("P(X=0|Y=1) =", 1 - EX_Y1)
print("P(X=1|Y=1) =", EX_Y1)

# ------------------------------------------------------------
# 6. Verify Expectation Formula
# ------------------------------------------------------------

print("\nVerification of:")
print("E[X|Y=y] = Σ x * p(x|y)")

# y = 0
verify0 = 0 * (1 - EX_Y0) + 1 * EX_Y0

# y = 1
verify1 = 0 * (1 - EX_Y1) + 1 * EX_Y1

print("\nFor y = 0:", verify0)
print("For y = 1:", verify1)

# ------------------------------------------------------------
# 7. Mean Square Error (MSE)
# ------------------------------------------------------------

# Variance for Bernoulli:
# Var(X) = p(1-p)

var0 = EX_Y0 * (1 - EX_Y0)
var1 = EX_Y1 * (1 - EX_Y1)

MSE = P_Y0 * var0 + P_Y1 * var1

print("\nMean Square Error Using E[X|Y]:")
print("MSE =", MSE)

# ------------------------------------------------------------
# 8. Global Mean Estimator
# ------------------------------------------------------------

EX = sum(X[w] * P[w] for w in Omega)

MSE_global = sum((X[w] - EX) ** 2 * P[w] for w in Omega)

print("\nGlobal Mean:")
print("E[X] =", EX)

print("\nMSE Using Global Average Only:")
print("MSE =", MSE_global)

# ------------------------------------------------------------
# 9. Law of Total Variance
# ------------------------------------------------------------

# Var(X)
Var_X = EX * (1 - EX)

# Var(E[X|Y])
Var_EXY = (
    P_Y0 * (EX_Y0 - EX) ** 2
    +
    P_Y1 * (EX_Y1 - EX) ** 2
)

print("\nLaw of Total Variance:")
print("Var(X) =", Var_X)
print("E[Var(X|Y)] =", MSE)
print("Var(E[X|Y]) =", Var_EXY)

print("\nCheck:")
print("E[Var(X|Y)] + Var(E[X|Y]) =",
      MSE + Var_EXY)

# ------------------------------------------------------------
# 10. Information Theory
# ------------------------------------------------------------

def entropy(probabilities):
    """
    Compute entropy in bits.
    """
    h = 0

    for p in probabilities:
        if p > 0:
            h -= p * math.log2(p)

    return h

# Entropy H(X)
HX = entropy([0.7, 0.3])

print("\nEntropy H(X):")
print("H(X) =", HX, "bits")

# Conditional Entropy H(X|Y)

H_X_given_Y0 = entropy([5/6, 1/6])
H_X_given_Y1 = entropy([1/2, 1/2])

H_X_given_Y = (
    P_Y0 * H_X_given_Y0
    +
    P_Y1 * H_X_given_Y1
)

print("\nConditional Entropy H(X|Y):")
print("H(X|Y) =", H_X_given_Y, "bits")

# Mutual Information
I_XY = HX - H_X_given_Y

print("\nMutual Information I(X;Y):")
print("I(X;Y) =", I_XY, "bits")

# Percentage uncertainty removed
percentage = (I_XY / HX) * 100

print("\nPercentage of Uncertainty Removed:")
print(percentage, "%")

# ------------------------------------------------------------
# 11. Joint Distribution Table
# ------------------------------------------------------------

print("\nJoint Distribution P(X,Y):")

joint = {
    (0, 0): P['w1'],
    (1, 0): P['w2'],
    (0, 1): P['w3'],
    (1, 1): P['w4']
}

for key, value in joint.items():
    print(f"P(X={key[0]}, Y={key[1]}) = {value}")

```

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

```python
# ============================================================
# Conditional Distribution of Multivariate Normal Distribution
# ============================================================

# This code demonstrates:
#
# 1. Construction of a joint multivariate normal distribution
# 2. Partitioning covariance matrices
# 3. Computing conditional mean
# 4. Computing conditional covariance
# 5. Evaluating the conditional Gaussian density
#
# Formula:
#
# X | Y=y ~ N(mu*, Sigma_X|Y)
#
# where
#
# mu* = mu_x + Sigma_XY Sigma_YY^{-1} (y - mu_y)
#
# Sigma_X|Y
# = Sigma_XX - Sigma_XY Sigma_YY^{-1} Sigma_YX
#
# ============================================================

import numpy as np



# Dimension of X
n = 2

# Dimension of Y
m = 1



# Mean of X
mu_x = np.array([
    [1.0],
    [2.0]
])

# Mean of Y
mu_y = np.array([
    [0.5]
])

# Joint mean vector
mu = np.vstack((mu_x, mu_y))

print("Joint Mean Vector:")
print(mu)

# ------------------------------------------------------------
# 3. Define Covariance Matrix Blocks
# ------------------------------------------------------------

# Covariance of X
Sigma_XX = np.array([
    [4.0, 1.5],
    [1.5, 3.0]
])

# Covariance of Y
Sigma_YY = np.array([
    [2.0]
])

# Cross covariance
Sigma_XY = np.array([
    [1.0],
    [0.5]
])

# Sigma_YX = transpose of Sigma_XY
Sigma_YX = Sigma_XY.T

# ------------------------------------------------------------
# 4. Build Joint Covariance Matrix
# ------------------------------------------------------------

Sigma = np.block([
    [Sigma_XX, Sigma_XY],
    [Sigma_YX, Sigma_YY]
])

print("\nJoint Covariance Matrix:")
print(Sigma)

# ------------------------------------------------------------
# 5. Check Positive Definiteness
# ------------------------------------------------------------

eigenvalues = np.linalg.eigvals(Sigma)

print("\nEigenvalues of Sigma:")
print(eigenvalues)

if np.all(eigenvalues > 0):
    print("\nSigma is positive definite.")
else:
    print("\nSigma is NOT positive definite.")

# ------------------------------------------------------------
# 6. Given Observation y
# ------------------------------------------------------------

y = np.array([
    [1.2]
])

print("\nObserved y:")
print(y)

# ------------------------------------------------------------
# 7. Compute Conditional Mean
# ------------------------------------------------------------

Sigma_YY_inv = np.linalg.inv(Sigma_YY)

mu_star = (
    mu_x
    +
    Sigma_XY @ Sigma_YY_inv @ (y - mu_y)
)

print("\nConditional Mean mu*:")
print(mu_star)

# ------------------------------------------------------------
# 8. Compute Conditional Covariance
# ------------------------------------------------------------

Sigma_conditional = (
    Sigma_XX
    -
    Sigma_XY @ Sigma_YY_inv @ Sigma_YX
)

print("\nConditional Covariance Sigma_(X|Y):")
print(Sigma_conditional)

# ------------------------------------------------------------
# 9. Determinant of Conditional Covariance
# ------------------------------------------------------------

det_conditional = np.linalg.det(Sigma_conditional)

print("\nDeterminant of Conditional Covariance:")
print(det_conditional)

# ------------------------------------------------------------
# 10. Define Conditional Gaussian Density Function
# ------------------------------------------------------------

def conditional_density(x, mu_star, Sigma_conditional):
    """
    Compute conditional multivariate normal density.

    Parameters:
        x : column vector
        mu_star : conditional mean
        Sigma_conditional : conditional covariance

    Returns:
        density value
    """

    n = len(x)

    Sigma_inv = np.linalg.inv(Sigma_conditional)

    det_sigma = np.linalg.det(Sigma_conditional)

    diff = x - mu_star

    exponent = (
        -0.5
        *
        (
            diff.T
            @ Sigma_inv
            @ diff
        )
    )

    normalization = (
        1
        /
        np.sqrt(
            (2 * np.pi) ** n
            *
            det_sigma
        )
    )

    density = normalization * np.exp(exponent)

    return density.item()

# ------------------------------------------------------------
# 11. Evaluate Density at Example Point
# ------------------------------------------------------------

x = np.array([
    [1.5],
    [2.2]
])

density_value = conditional_density(
    x,
    mu_star,
    Sigma_conditional
)

print("\nPoint x:")
print(x)

print("\nConditional Density f(X|Y=y):")
print(density_value)

# ------------------------------------------------------------
# 12. Verify Matrix Formula Components
# ------------------------------------------------------------

print("\nVerification of Formula Components")

print("\nSigma_XY:")
print(Sigma_XY)

print("\nSigma_YY^{-1}:")
print(Sigma_YY_inv)

print("\n(y - mu_y):")
print(y - mu_y)

print("\nSigma_XY Sigma_YY^{-1} (y - mu_y):")
print(
    Sigma_XY
    @ Sigma_YY_inv
    @ (y - mu_y)
)

# ------------------------------------------------------------
# 13. Final Theoretical Statement
# ------------------------------------------------------------

print("\n================================================")
print("FINAL RESULT")
print("================================================")

print("\nX | Y=y follows a multivariate normal distribution:")

print("\nX | Y=y ~ N(mu*, Sigma_(X|Y))")

print("\nwhere")

print("\nmu* =")
print(mu_star)

print("\nSigma_(X|Y) =")
print(Sigma_conditional)

```

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**
1.  **Hidden Temperature ($X$):** The standardized temperature deviation follows a standard normal distribution:
    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

---

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.
* Find the covariance $\text{Cov}(X, Y)$.

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.
* Using the property for joint normals: $E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$.
* Express this as a "signal-to-total-variance" ratio.

**4. Verification of the Residual**
Let $\widehat{X} = E[X \mid Y]$. Define the estimation error (residual) as $Z = X - \widehat{X}$.
* Prove that $E[Z] = 0$.
* Verify that the error $Z$ is uncorrelated with the observation $Y$ (the Orthogonality Principle).

**5. Information Theory: Differential Entropy**
* Compute the differential entropy $h(X)$.
* Compute the conditional differential entropy $h(X \mid Y)$.
* *Note: The variance of the conditional distribution $(X \mid Y=y)$ is given by $\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)}$.*

**6. Mutual Information**
Calculate $I(X; Y) = h(X) - h(X \mid Y)$.
* What percentage of the "uncertainty" about the temperature is removed by observing the power meter?
* How would this change if the mechanical noise $Q$ had a variance of $100$ instead of $4$?

**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*
Using your result from Question 3, explain why $Y$ is a **biased** estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).



```python

# CNC SPINDLE MONITORING : CONDITIONAL EXPECTATION

import numpy as np
import math

# GIVEN DATA

# X ~ N(0,1)
mu_X = 0
var_X = 1

# Q ~ N(0,4)
mu_Q = 0
var_Q = 4

# Y = X + Q


print("GIVEN")


print("X ~ N(0,1)")
print("Q ~ N(0,4)")
print("Y = X + Q")

# QUESTION 1 : JOINT DISTRIBUTION


print("QUESTION 1 : JOINT DISTRIBUTION")


# Mean of Y

mu_Y = mu_X + mu_Q

print("\nMean of Y:")

print("E[Y]")
print("= E[X+Q]")
print("= E[X] + E[Q]")
print("=", mu_X, "+", mu_Q)
print("=", mu_Y)

# Variance of Y

var_Y = var_X + var_Q

print("\nVariance of Y:")

print("Var(Y)")
print("= Var(X+Q)")

print("\nSince X and Q are independent:")

print("Var(Y)")
print("= Var(X) + Var(Q)")
print("=", var_X, "+", var_Q)
print("=", var_Y)

print("\nTherefore:")

print("Y ~ N(0,5)")

# Covariance
cov_XY = var_X

print("\nCovariance Between X and Y:")

print("Cov(X,Y)")
print("= Cov(X , X+Q)")
print("= Cov(X,X) + Cov(X,Q)")

print("\nSince X and Q are independent:")

print("Cov(X,Q)=0")

print("\nTherefore:")

print("Cov(X,Y)")
print("= Var(X)")
print("=", cov_XY)

# Covariance matrix

Sigma = np.array([
    [var_X, cov_XY],
    [cov_XY, var_Y]
])

print("\nJoint Covariance Matrix:")

print(Sigma)

# QUESTION 2 : INFORMATION GEOMETRY


print("QUESTION 2 : INFORMATION GEOMETRY")

print("""
X is NOT sigma(Y)-measurable.

Reason:

Y = X + Q

The observation Y contains:
1) temperature contribution X
2) mechanical noise Q

Therefore one observed value of Y
cannot uniquely determine X.
""")

print("Example:")

print("""
Suppose:

Y = 5

Case 1:
X = 4
Q = 1

Then:
Y = 4 + 1 = 5

Case 2:
X = 1
Q = 4

Then:
Y = 1 + 4 = 5
""")

print("""
Therefore the engineer cannot know whether:
- the spindle is overheating
OR
- the cutting load is large.
""")

# QUESTION 3 : CONDITIONAL EXPECTATION


print("QUESTION 3 : CONDITIONAL EXPECTATION")


gain = cov_XY / var_Y

print("\nFormula:")

print("""
E[X|Y=y]
= mu_X + Cov(X,Y)/Var(Y) * (y-mu_Y)
""")

print("Substituting values:")

print("""
mu_X = 0
mu_Y = 0

Cov(X,Y) = 1
Var(Y) = 5
""")

print("Therefore:")

print("E[X|Y=y]")
print("= (1/5)y")

print("\nHence:")

print("E[X|Y] = (1/5)Y")

# Signal-to-total-variance ratio

print("\nSignal-to-total-variance ratio:")

print("""
E[X|Y]
=
Var(X) / [Var(X)+Var(Q)] * Y
""")

print("=")
print(var_X, "/", f"({var_X}+{var_Q})", "* Y")

print("=")
print(gain, "* Y")

# Example

y_example = 10

estimate = gain * y_example

print("\nExample:")

print("If Y = 10")

print("Best estimate:")
print("E[X|Y=10]")
print("=", gain, "*", y_example)
print("=", estimate)

# QUESTION 4 : RESIDUAL ERROR
print("QUESTION 4 : RESIDUAL ERROR")


print("""
Define:

X_hat = E[X|Y]
= (1/5)Y

Residual:

Z = X - X_hat
""")
# Mean of residual

EZ = mu_X - gain * mu_Y

print("Part (a) : Show E[Z]=0")

print("\nE[Z]")
print("= E[X-(1/5)Y]")
print("= E[X] - (1/5)E[Y]")

print("=")
print(mu_X, "-", gain, "*", mu_Y)

print("=")
print(EZ)

# Orthogonality principle

cov_ZY = cov_XY - gain * var_Y

print("\nPart (b) : Orthogonality Principle")

print("\nCov(Z,Y)")
print("= Cov(X-(1/5)Y , Y)")

print("\nUsing linearity:")

print("= Cov(X,Y) - (1/5)Var(Y)")

print("=")
print(cov_XY, "-", gain, "*", var_Y)

print("=")
print(cov_ZY)

print("\nTherefore residual is uncorrelated with Y.")

# QUESTION 5 : DIFFERENTIAL ENTROPY


print("QUESTION 5 : DIFFERENTIAL ENTROPY")


def gaussian_entropy(variance):

    return 0.5 * math.log(2 * math.pi * math.e * variance)

# h(X)

h_X = gaussian_entropy(var_X)

print("\nEntropy Formula:")

print("""
h(X)
= (1/2)log(2*pi*e*sigma^2)
""")

print("For X:")

print("sigma_X^2 =", var_X)

print("\nh(X)")
print("=", h_X, "nats")
# Conditional variance

var_X_given_Y = (
    var_X
    -
    (cov_XY ** 2) / var_Y
)

print("\nConditional variance:")

print("""
Var(X|Y)
=
Var(X) - Cov(X,Y)^2 / Var(Y)
""")

print("=")
print(var_X, "-", f"({cov_XY}^2)/{var_Y}")

print("=")
print(var_X_given_Y)

# Conditional entropy

h_X_given_Y = gaussian_entropy(var_X_given_Y)

print("\nConditional entropy:")

print("h(X|Y)")
print("=", h_X_given_Y, "nats")

# QUESTION 6 : MUTUAL INFORMATION



print("QUESTION 6 : MUTUAL INFORMATION")


I_XY = h_X - h_X_given_Y

print("\nMutual Information Formula:")

print("""
I(X;Y)
= h(X) - h(X|Y)
""")

print("\nSubstituting values:")

print("=")
print(h_X, "-", h_X_given_Y)

print("=")
print(I_XY, "nats")

# Percentage removed


percentage = (I_XY / h_X) * 100

print("\nPercentage uncertainty removed:")

print("=")
print(f"({I_XY}/{h_X}) * 100")

print("=")
print(percentage, "%")

# Noise variance = 100

print("\nCASE : Noise variance becomes 100")

var_Q_new = 100

var_Y_new = var_X + var_Q_new

gain_new = var_X / var_Y_new

print("\nNew variance of Y:")

print("Var(Y)")
print("=", var_X, "+", var_Q_new)
print("=", var_Y_new)

print("\nNew conditional expectation:")

print("E[X|Y]")
print("=", gain_new, "* Y")

print("""
Interpretation:

When noise variance becomes extremely large,
the observation Y becomes almost useless
for estimating X.
""")

# QUESTION 7 : DIAGNOSTIC INTERPRETATION


print("QUESTION 7 : DIAGNOSTIC INTERPRETATION")


print("""
Manager says:

'Best estimate should be Y itself.'

This is incorrect because:

Y = X + Q

The observation contains noise Q.
""")

print("""
Optimal estimator:

E[X|Y]
= (1/5)Y
""")

print("""
This SHRINKS observations toward zero.
""")

print("Example:")

print("""
If:

Y = 10

Direct estimate:
X_hat = 10

Optimal estimate:
X_hat = 2
""")

print("""
Reason:

Most variation in Y
comes from noise:

Var(Q)=4

while signal variance is only:

Var(X)=1
""")

print("""
Therefore large Y values are often
caused mostly by noise.

This phenomenon is called:

SHRINKAGE
""")

print("""
Applications:
- Kalman filtering
- Bayesian estimation
- Wiener filtering
- Signal processing
- Machine learning
""")


```
